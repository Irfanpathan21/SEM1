# 1. Simple Interest Program

### Algorithm
1. Start
2. Read principal (P), rate of interest (R), and number of years (T)
3. Calculate simple interest using the formula:
   `SI = (P * R * T) / 100`
4. Print the simple interest
5. End


```python
# Simple Interest Program in C
#include <stdio.h>

int main() {
    float principal, rate, time, simpleInterest;

    // Input from the user
    printf("Enter Principal: ");
    scanf("%f", &principal);
    printf("Enter Rate of Interest: ");
    scanf("%f", &rate);
    printf("Enter Time (in years): ");
    scanf("%f", &time);

    // Calculate Simple Interest
    simpleInterest = (principal * rate * time) / 100;

    // Output the result
    printf("Simple Interest: %.2f\n", simpleInterest);

    return 0;
}
```

# 2. Greatest of Three Numbers Using Conditional Operator

### Algorithm
1. Start
2. Read three numbers: a, b, c
3. Use the conditional operator to find the greatest number:
   - `greatest = (a > b) ? (a > c ? a : c) : (b > c ? b : c)`
4. Print the greatest number
5. End


```python
# Greatest of Three Numbers Using Conditional Operator in C
#include <stdio.h>

int main() {
    int a, b, c;
    int greatest;

    // Input from the user
    printf("Enter three numbers: ");
    scanf("%d %d %d", &a, &b, &c);

    // Find greatest using conditional operator
    greatest = (a > b) ? (a > c ? a : c) : (b > c ? b : c);

    // Output the result
    printf("The greatest number is: %d\n", greatest);

    return 0;
}
```

# 3. Leap Year Check Program

### Algorithm
1. Start
2. Read the year (Y) from the user
3. If (Y % 4 == 0) and (Y % 100 != 0) or (Y % 400 == 0), then it's a leap year.
4. Print if the year is a leap year or not
5. End


```python
# Leap Year Check Program in C
#include <stdio.h>

int main() {
    int year;

    // Input from the user
    printf("Enter year: ");
    scanf("%d", &year);

    // Leap year check
    if ((year % 4 == 0 && year % 100 != 0) || (year % 400 == 0)) {
        printf("%d is a leap year.\n", year);
    } else {
        printf("%d is not a leap year.\n", year);
    }

    return 0;
}
```

# 4. Roots of a Quadratic Equation

### Algorithm
1. Start
2. Read the coefficients a, b, and c of the quadratic equation `ax^2 + bx + c = 0`
3. Calculate the discriminant: `D = b^2 - 4ac`
4. If `D > 0`, two real and distinct roots exist:
   - `root1 = (-b + sqrt(D)) / 2a`
   - `root2 = (-b - sqrt(D)) / 2a`
5. If `D == 0`, one real root exists:
   - `root = -b / 2a`
6. If `D < 0`, print that there are no real roots
7. End


```python
# Roots of a Quadratic Equation in C
#include <stdio.h>
#include <math.h>

int main() {
    float a, b, c, discriminant, root1, root2;

    // Input from the user
    printf("Enter coefficients a, b, c: ");
    scanf("%f %f %f", &a, &b, &c);

    // Calculate discriminant
    discriminant = b*b - 4*a*c;

    // Check the discriminant and calculate roots
    if (discriminant > 0) {
        root1 = (-b + sqrt(discriminant)) / (2 * a);
        root2 = (-b - sqrt(discriminant)) / (2 * a);
        printf("Roots are real and distinct: %.2f and %.2f\n", root1, root2);
    } else if (discriminant == 0) {
        root1 = -b / (2 * a);
        printf("Root is real and same: %.2f\n", root1);
    } else {
        printf("No real roots\n");
    }

    return 0;
}
```

# 5. Menu-driven Program for Arithmetic Operations

### Algorithm
1. Start
2. Display the menu: 1. Add 2. Subtract 3. Multiply 4. Divide
3. Read the user's choice
4. Read two numbers (a, b)
5. Perform the operation based on the choice:
   - If 1, add the numbers
   - If 2, subtract the numbers
   - If 3, multiply the numbers
   - If 4, divide the numbers
6. Print the result
7. End


```python
# Menu-driven Program for Arithmetic Operations in C
#include <stdio.h>

int main() {
    int choice;
    float num1, num2, result;

    // Display menu
    printf("Menu:\n");
    printf("1. Add\n2. Subtract\n3. Multiply\n4. Divide\n");
    printf("Enter your choice: ");
    scanf("%d", &choice);

    // Input two numbers
    printf("Enter two numbers: ");
    scanf("%f %f", &num1, &num2);

    // Perform the chosen operation
    switch (choice) {
        case 1:
            result = num1 + num2;
            printf("Result: %.2f\n", result);
            break;
        case 2:
            result = num1 - num2;
            printf("Result: %.2f\n", result);
            break;
        case 3:
            result = num1 * num2;
            printf("Result: %.2f\n", result);
            break;
        case 4:
            if (num2 != 0) {
                result = num1 / num2;
                printf("Result: %.2f\n", result);
            } else {
                printf("Cannot divide by zero\n");
            }
            break;
        default:
            printf("Invalid choice\n");
    }

    return 0;
}
```
